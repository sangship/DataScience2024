```python
# # df에서 결측치가 있는 행 삭제하려는데, 그전에 별개의 df_tmp 복사해서 확인해보기
# df_tmp = df.copy()
# df_tmp.head()

# df_drop = df_tmp.dropna()
# df_drop.head()

# # df_drop의 행 개수와 df_tmp의 행 개수를 비교
# print(len(df_drop))
# print(len(df_tmp))
# print(len(df_drop) - len(df_tmp))


# df에서 결측치가 있는 행 삭제
df_drop = df.dropna()

# 결과 확인
print(len(df_drop))
print(len(df))
print(len(df_drop) - len(df))

```

```python
df_drop.head()
```

```python
# df_drop의 일시 칼럼을 이용해서, 년도별로 (평균기온, 평균최저기온, 평균최고기온)을 계산해서 새로운 dataframe에 넣기

# '일시' 컬럼에서 년도 추출
df_drop.loc[:, '년도'] = df_drop['일시'].str[:4]

# 년도별 평균 기온 계산
df_yearly = df_drop.groupby('년도').agg({'평균기온': 'mean', '최저기온': 'mean', '최고기온': 'mean'})

# 컬럼명 변경
df_yearly.columns = ['평균기온', '평균최저기온', '평균최고기온']

# 결과 확인
df_yearly.head()

```

```python
# df_yearly을 이용해서 년도별로 평균기온, 평균최저기온, 평균최고기온 그래프 그리기

import matplotlib.pyplot as plt

# 그래프 그리기
plt.plot(df_yearly.index, df_yearly['평균기온'], label='평균기온')
plt.plot(df_yearly.index, df_yearly['평균최저기온'], label='평균최저기온')
plt.plot(df_yearly.index, df_yearly['평균최고기온'], label='평균최고기온')

# 그래프 설정
plt.xlabel('년도')
plt.ylabel('기온(℃)')
plt.title('년도별 평균 기온')
plt.legend()
plt.grid(True)

# 그래프 출력
plt.show()
```

```python

# df_yearly에서 평균기온, 평균최저기온, 평균최고기온 값이 급락하는 구간


# # 급락 기준 설정 (예: 전년 대비 5도 이상 감소)
# threshold = -5

# # 급락 구간 찾기
# result = []
# for col in ['평균기온', '평균최저기온', '평균최고기온']:
#     temp_diff = df_yearly[col].diff()
#     drop_indices = temp_diff[temp_diff < threshold].index
#     for idx in drop_indices:
#         result.append((idx, col, df_yearly.loc[idx, col] - df_yearly.loc[str(int(idx) - 1), col]))

# # 결과 출력
# for year, col, diff in result:
#     print(f"{year}년 {col} 급락: {diff:.2f}℃ 감소")




# # df_yearly에서 평균기온, 평균최저기온, 평균최고기온 값이 급락하는 구간

# # 급락 기준 설정 (예: 전년 대비 1도 이상 감소)
# threshold = -1

# # 급락 구간 찾기
# result = []
# for col in ['평균기온', '평균최저기온', '평균최고기온']:
#     temp_diff = df_yearly[col].diff()
#     drop_indices = temp_diff[temp_diff < threshold].index
#     for idx in drop_indices:
#         previous_year = idx - 1
#         if previous_year in df_yearly.index:
#             result.append((idx, col, df_yearly.loc[idx, col] - df_yearly.loc[previous_year, col]))

# # 결과 출력
# for year, col, diff in result:
#     print(f"{year}년 {col} 급락: {diff:.2f}℃ 감소")


# 인덱스를 숫자로 변환
df_yearly.index = df_yearly.index.astype(int)

# 급락 기준 설정 (예: 전년 대비 1도 이상 감소)
threshold = -1

# 급락 구간 찾기
result = []
for col in ['평균기온', '평균최저기온', '평균최고기온']:
    temp_diff = df_yearly[col].diff()
    drop_indices = temp_diff[temp_diff < threshold].index
    for idx in drop_indices:
        previous_year = idx - 1
        if previous_year in df_yearly.index:
            result.append((idx, col, df_yearly.loc[idx, col] - df_yearly.loc[previous_year, col]))

# 결과 출력
for year, col, diff in result:
    print(f"{year}년 {col} 급락: {diff:.2f}℃ 감소")





```

```python
#10년 주기로 년도별 평균기온, 평균최저기온, 평균최고기온 그래프 그리기.
#여기서 10년 주기라 함은 1921부터 1930, 1931부터 1940 이렇게,
# df_yearly의 데이터값의 첫번째 행부터 시작

# 첫 번째 년도 추출
start_year = int(df_yearly.index[0])

# 10년 단위 기간 설정
periods = [(str(year), str(year + 9)) for year in range(start_year, int(df_yearly.index[-1]) + 1, 10)]

# 각 기간별 평균 기온 계산
result = []
for start, end in periods:
    df_period = df_yearly[(df_yearly.index >= start) & (df_yearly.index <= end)]
    if not df_period.empty:
        mean_temp = df_period['평균기온'].mean()
        mean_low = df_period['평균최저기온'].mean()
        mean_high = df_period['평균최고기온'].mean()
        result.append([start + '-' + end, mean_temp, mean_low, mean_high])

# 결과를 DataFrame으로 변환
df_result = pd.DataFrame(result, columns=['기간', '평균기온', '평균최저기온', '평균최고기온'])

# 그래프 그리기
plt.plot(df_result['기간'], df_result['평균기온'], label='평균기온')
plt.plot(df_result['기간'], df_result['평균최저기온'], label='평균최저기온')
plt.plot(df_result['기간'], df_result['평균최고기온'], label='평균최고기온')

# 그래프 설정
plt.xlabel('기간')
plt.ylabel('기온(℃)')
plt.title('10년 단위 기간별 평균 기온')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

# 그래프 출력
plt.show()
```
